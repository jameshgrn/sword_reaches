```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os

df = pd.read_excel('/Users/jakegearon/PycharmProjects/NatureGeoSubmission/rebirth/data_to_submit.xlsx')
df.dropna(subset=['AR_FULL_W', 'width'], inplace=True)
df.columns
```

```python
sns.scatterplot(x=df['normalized_dist']/df['parent_ar_slope_mean'], y=df['AR_FULL_W']/df['width'], data=df, hue=df['superelevation_ov_mean']*df['XS/DS'], palette='viridis')
plt.axhline(y=np.median(df['AR_FULL_W']/df['width']), color='r', linestyle='-')
#plt.loglog()
print(np.median(df['AR_FULL_W']/df['width']))
plt.xscale('log')
plt.show()
```

```python
# Calculate the log-transformed ratio
log_ratio = np.log((df['AR_FULL_W'] - df['width']) / df['width'])

# Filter out non-positive values
positive_log_ratio = log_ratio[log_ratio > 0]

# Plot the distribution of the positive log-transformed ratio
sns.displot(positive_log_ratio, bins=20)
plt.xlabel('(AR width - channel width)/(channel width)')  # Added label for x-axis
plt.ylabel('Frequency')  # Added label for y-axis

# Draw a vertical line at the median of the positive log-transformed ratio
plt.axvline(x=np.median(positive_log_ratio), color='r', linestyle='-')
plt.show()
```
